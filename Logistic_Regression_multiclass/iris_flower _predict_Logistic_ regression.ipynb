<h2 style='color:green' align='center'> Logistic Regression: Multiclass Classification </h2>

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

irys = load_iris()
```

```python
irys.data
```

```python
irys.target
```

```python
irys.target_names
```

<h4 style='color:blue'>Create and train logistic regression model</h4>

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(irys.data,irys.target,test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

<h4 style='color:blue'> Measure accuracy of the Model<h4>

```python
model.score(X_test,y_test)
```

<h4 style='color:blue'>Confussion Matrix</h4>

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (6,4))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

We can observe that 2 times the model predicted "virginica" when it was "versicolor"
