```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

We create a dataframe with the information of the csv file canada_per_capita_income

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

We create a plot to visualize the data of the csv file.

```python
%matplotlib inline
plt.xlabel('Year',fontsize=20)
plt.ylabel('income per capita (US$)',fontsize=20)
plt.scatter(df['year'], df['per capita income (US$)'])
```

We use linear regression to predict the income per capita of the year 2025 

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
reg.predict([[2025]])
```

Using linear regression to predict canada's adjusted income per capita (US$) of the year 2025 we get the value of $44652.0181496

We can also plot our prediction values to have a visual comparison between the real values and the values given by our Linear regression model.

```python
%matplotlib inline
plt.xlabel('Year',fontsize=20)
plt.ylabel('income per capita (US$)',fontsize=20)
plt.scatter(df['year'], df['per capita income (US$)'])
plt.plot(df['year'],reg.predict(df[['year']]),color='red')
```
